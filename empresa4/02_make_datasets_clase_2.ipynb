```python
df['periodo'] = df['periodo'].astype(int)
df['product_id'] = df['product_id'].astype(int)
df['customer_id'] = df['customer_id'].astype(int)
df['plan_precios_cuidados'] = df['plan_precios_cuidados'].astype(int)
df['cust_request_qty'] = df['cust_request_qty'].astype(float)
df['cust_request_tn'] = df['cust_request_tn'].astype(float)
df['tn'] = df['tn'].astype(float)
```

```python
print(df["cust_request_qty"].sum(), df["cust_request_tn"].sum(), df["tn"].sum())
```

```python
# print unique number of customer_id, product_id, and period
print(df["customer_id"].nunique(), df["product_id"].nunique(), df["periodo"].nunique())
```

```python
df_clientes = df.groupby(['periodo', 'customer_id', 'product_id'])[['cust_request_qty', 'cust_request_tn', 'tn']].sum()
df_productos = df.groupby(['periodo', 'product_id', ])[['cust_request_qty', 'cust_request_tn', 'tn']].sum()
```

```python
print("ORIGINAL", df["cust_request_qty"].sum(), df["cust_request_tn"].sum(), df["tn"].sum())
print("CLIENTES", df_clientes["cust_request_qty"].sum(), df_clientes["cust_request_tn"].sum(), df_clientes["tn"].sum())
print("PRODUCTOS", df_productos["cust_request_qty"].sum(), df_productos["cust_request_tn"].sum(), df_productos["tn"].sum())
```

```python
df_productos.reset_index(inplace=True)
df_clientes.reset_index(inplace=True)
```

```python
product_ids_maestro = original_categories["product_id"].unique().tolist()
product_ids_sellout = df_productos["product_id"].unique().tolist()
faltantes_en_sellout = list(set(product_ids_maestro) - set(product_ids_sellout))
faltantes_en_maestro = list(set(product_ids_sellout) - set(product_ids_maestro))
len(faltantes_en_sellout), len(faltantes_en_maestro)
```

```python
# for each value in faltantes_en_sellout, add a row to df_productos with 0 values and periodo = 201701
for product_id in faltantes_en_sellout:
    df_productos = pd.concat([df_productos, pd.DataFrame([{'product_id': product_id, 'periodo': 201701, 'cust_request_qty': 0, 'cust_request_tn': 0, 'tn': 0}])], axis=0, ignore_index=True)
    df_clientes = pd.concat([df_clientes, pd.DataFrame([{'product_id': product_id, 'periodo': 201701, 'customer_id': 10001, 'cust_request_qty': 0, 'cust_request_tn': 0, 'tn': 0}])], axis=0, ignore_index=True)
df_productos["product_id"].nunique(), df_clientes["product_id"].nunique()
```

```python
# Find all unique values for 'periodo' and 'product_id'
unique_periodos = df['periodo'].unique()
unique_product_ids = df_productos['product_id'].unique()
unique_customer_ids = df['customer_id'].unique()
len(unique_periodos), len(unique_product_ids), len(unique_customer_ids)
```

```python
print("Shape antes: ", df_productos.shape)

from itertools import product

# Create all possible combinations of 'periodo' and 'product_id'
all_combinations = list(product(unique_periodos, unique_product_ids))

# Convert the combinations into a DataFrame
all_combinations_df = pd.DataFrame(all_combinations, columns=['periodo', 'product_id'])

# Merge with the existing DataFrame to find missing rows
combined_df = all_combinations_df.merge(df_productos, on=['periodo', 'product_id'], how='left')

# Identify the missing rows by looking for NaNs
missing_rows = combined_df[combined_df['cust_request_qty'].isna()]

# For the missing rows, set 'cust_request_qty', 'cust_request_tn', and 'tn' to 0
# We also need to decide how to handle 'product_category'. For now, we'll set it as 'Unknown'
missing_rows.loc[:, 'cust_request_qty'] = 0
missing_rows.loc[:, 'cust_request_tn'] = 0
missing_rows.loc[:, 'tn'] = 0

# Now we append the missing rows to the original DataFrame
# convert to pd.concat: complete_df = df_productos.append(missing_rows[df_productos.columns], ignore_index=True)
df_productos = pd.concat([df_productos, missing_rows[df_productos.columns]], ignore_index=True)

# Return the number of new rows added and the path to the new file
num_new_rows = len(missing_rows)
print(num_new_rows)

print("Shape Después: ", df_productos.shape)
```

```python
# Debería tener resto 0
print(df_productos.shape[0] / 36)
```

```python
print("Shape antes: ", df_clientes.shape)
print(df_clientes.shape[0] / 36)

from itertools import product

# Create all possible combinations of 'periodo' and 'product_id'
all_combinations = list(product(unique_periodos, unique_customer_ids, unique_product_ids))

# Convert the combinations into a DataFrame
all_combinations_df = pd.DataFrame(all_combinations, columns=['periodo', 'customer_id', 'product_id'])

# Merge with the existing DataFrame to find missing rows
combined_df = all_combinations_df.merge(df_clientes, on=['periodo', 'customer_id', 'product_id'], how='left')

# Identify the missing rows by looking for NaNs
missing_rows = combined_df[combined_df['cust_request_qty'].isna()]

# For the missing rows, set 'cust_request_qty', 'cust_request_tn', and 'tn' to 0
# We also need to decide how to handle 'product_category'. For now, we'll set it as 'Unknown'
missing_rows.loc[:, 'cust_request_qty'] = 0
missing_rows.loc[:, 'cust_request_tn'] = 0
missing_rows.loc[:, 'tn'] = 0

# Now we append the missing rows to the original DataFrame
# convert to pd.concat: complete_df = df_productos.append(missing_rows[df_productos.columns], ignore_index=True)
df_clientes = pd.concat([df_clientes, missing_rows[df_clientes.columns]], ignore_index=True)

# Return the number of new rows added and the path to the new file
num_new_rows = len(missing_rows)
print(num_new_rows)

print("Shape Después: ", df_clientes.shape)

# Debería tener resto 0
print(df_clientes.shape[0] / 36)
```

```python
df_productos = pd.merge(df_productos, original_categories[['product_id', 'cat1', 'cat2', 'sku_size']], on='product_id', how='left')
df_productos.rename(columns={'cat1': 'product_category'}, inplace=True)
df_productos["product_category"].fillna('unknown', inplace=True)
df_productos["cat2"].fillna('unknown', inplace=True)
df_productos["sku_size"].fillna(0, inplace=True)

df_clientes = pd.merge(df_clientes, original_categories[['product_id', 'cat1', 'cat2', 'sku_size']], on='product_id', how='left')
df_clientes.rename(columns={'cat1': 'product_category'}, inplace=True)
df_clientes["product_category"].fillna('unknown', inplace=True)  
df_clientes["cat2"].fillna('unknown', inplace=True)
df_clientes["sku_size"].fillna(0, inplace=True)
```
