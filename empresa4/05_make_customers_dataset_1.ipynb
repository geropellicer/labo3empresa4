```python
df_desglosado = get_dataset("02_productos_desglosado")
df_desglosado
```

```python
# group df_desglosado by "periodo" and "customer_id", sum cust_request_qty, cust_request_tn and tn columns and discard the rest
df_desglosado_agrupado = df_desglosado.groupby(["periodo", "customer_id"]).sum()[["cust_request_qty", "cust_request_tn", "tn"]]
df_desglosado_agrupado
```
